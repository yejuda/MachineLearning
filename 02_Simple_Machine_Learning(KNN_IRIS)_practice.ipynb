#### Data PreProcessing

##### 1. Split X and Y

```python
X_iris = iris[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
y_iris = iris['variety']
```

Column 명을 변경

```python
# 컬럼명 변경 -> 'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'
X_iris_columns = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
X_iris.columns = X_iris_columns
X_iris.head(5)
```

```python
y_iris
```

```python
y_iris.unique()  # 중복값 없이 출력   
```

```python
y_iris['variety'].unique()   # 위에 y_iris를 데이터 프레임으로 만들면 될 것임  # pd.DataFrame(~~~)
```

### 3. Split train set, test set

```python
from pandas.core.common import random_state
# 함수를 이용해서 나누기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_iris, y_iris, test_size = 0.3, random_state = 0)   # X는 데이터 y는 정답  # random_state = 0는 고정 시드값을 의미
```

### 4. Machine learning modeling

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)  # n_neighbors로 k를 정해줌/ 제일 가까운 것 뽑아내는 것

# n_neighbors는 k의 개수
```

### 5. Training

```python
X_train
```

```python
y_train
```

### 데이터를 fit시키기   
- 훈련 데이터셋으로부터 모델을 만들려면 knn객체의fit 메서드를 사용한다.    
- 훈련 세트를 사용하여 분류 모델을 학습시킨다. KNeifhborsClassifer에서의 학습은 예측할 때 이웃을 찾을 수 있도록 데이터를 저장하는 것이다. 

  - x데이터는 여러 개의 차원으로 이루어진 배열,   y데이터는 레이블(각 점들의 분류 결과)가 된다.   


```python
## knn 알고리즘 traiging
knn.fit(X_train, y_train)  
```

y_train = y_train.values.ravel() 의 코드는 (m,n)형태의 m*n개의 데이터를 한개의 (m*n,)시리즈 형태로 만드는 메소드입니다. input 데이터의 형식을 변경해주기위해서 작성하는 코드가 아닐까 생각합니다.

```python
print(y_train.shape)
print(y_train.values.ravel().shape)   # ravel() : 1차원으로 바꾸는 메소드 

# 참고 - flatten()으로 차원 바꾸는 법
# np.array(y_test).flatten()
```

```python
## knn 알고리즘 traiging
knn.fit(X_train, y_train.values.ravel())
```

### 6. Predict

```python
import numpy as np

# 'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)' 을 임의로 정한다
# 5.3, 2.1, 1.3, 0.2 의 수치로 정함
Our_X = np.array([[5.3, 2.1, 1.3, 0.2]])
Our_X
```

```python
# 임의의 데이터로 prediction
prediction = knn.predict(Our_X)
print("예측: {}".format(prediction))  
```

### 7. Accuracy

```python
y_pred = knn.predict(X_test)
print(y_pred.shape)
print("테스트 세트에 대한 예측값:\n {}".format(y_pred))
```

```python
# y_test.shape
```

```python
# flatten ->  y_test 를 1차원 matrix 로 변경
y_test_flatten = np.array(y_test).flatten()
```

```python
y_pred == y_test_flatten
```

```python
print("테스트 세트의 정확도: {:.2f}".format(np.mean(y_pred == y_test_flatten)))  # True = 1, False = 0  -> 이것들을 다 더해서 평균 낸 것이 정확도 0.98
```

```python
print("테스트 세트의 정확도: {:.2f}".format(knn.score(X_test, y_test_flatten)))
```

```python
np.mean(y_pred == y_test_flatten)
knn.score(X_test, y_test_flatten)
```
